## EDA

## Ploting graph to find number of Male and Female shoppers

```python
genderPlot=sb.countplot(x='Gender',data=dataframe)
```

```python
genderPlot=sb.countplot(x='Gender',data=dataframe)
for bars in genderPlot.containers:
    genderPlot.bar_label(bars)
```

```python
sales_purchase_classification_by_gender=dataframe.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Gender',y='Amount',data=sales_purchase_classification_by_gender)
```

## Occupation

```python
occupationPlot=sb.countplot(x='Occupation', data=dataframe)


for bars in occupationPlot.containers:
    occupationPlot.bar_label(bars)
```

```python
sales_occupation=dataframe.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Occupation' , y='Amount' , data=sales_occupation)
```

```python
sales_occupation=dataframe.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sb.barplot(x='Occupation' , y='Orders' , data=sales_occupation)
```

 ## Product Category

```python
product_category=sb.countplot(x='Product_Category' , data=dataframe)
for bars in product_category.containers:
    product_category.bar_label(bars)
```

```python
sales_occupation=dataframe.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sb.barplot(x='Product_Category' , y='Orders' , data=sales_occupation)
```

```python
sales_occupation=dataframe.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Product_Category' , y='Amount' , data=sales_occupation)
```

## Top selling products 

```python
top_selling_products=dataframe.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sb.barplot(x='Product_ID', y='Orders' ,data=top_selling_products)
```
